1 Introduction to Matplotlib
Your goal is to use matplotlib.pyplot and NumPy to plot several sine waves
by looping through amplitude and frequency values.
1. Write a function make sine wave(x, A, w) that returns:
y = A · sin(w · x)
where x is an array, A is amplitude, and w is angular frequency.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
def make_sine_wave(x, A, w):
    y = A*np.sin(w*x)
    return y
make_sine_wave(1, 2, np.pi)
```

2. Create an array x using NumPy that goes from 0 to 2π with 1000 points.

```python
x = np.arange(0,2*np.pi, np.pi/500)
x
```

3. Make two arrays:
• One with 5 amplitude values (e.g. [0.5, 1, 1.5, 2, 2.5]).
• One with 5 frequency values (e.g. [1, 2, 3, 4, 5]).

```python
amp = [1, np.pi, np.e, 4.2, 1/np.pi]
freq = [1, np.pi, np.e, 4.2, 1/np.pi]
```

4. Create a figure using plt.figure(). Give an appropriate value for the
figsize argument.
5. Loop through the amplitude and frequency arrays using zip(). For each
pair:
• Call your sine wave function to get y.
• Plot x vs. y on the same figure.
• Use a label to include A and w values in your legend.
6. Add a title, axis labels, legend, and grid. Then use plt.show() to display
the plot.
Bonus: Use different line styles for each sine wave.

```python
plt.figure(figsize=(6.4, 4.8))
# I could not figure out how to use zip for this question, since we did not learn it in class. Instead, I used this following way using a for loop.
for i in range(0,5):
    y = make_sine_wave(x, amp[i], freq[i])
    plt.plot(x,y, linestyle = (i, (i+1, abs(i-3))),label = f'A = {amp[i]}, w = {freq[i]}')
    plt.legend(bbox_to_anchor=(0.5, -0.3), loc='center', borderaxespad=0.)
    plt.grid("both")
    plt.ylabel('A * sin (w*x)')
    plt.xlabel('x')
    plt.title("Sine Waves")
plt.show()
```

3 Seaborn!
Use seaborn to explore the penguins dataset, which includes measurements of
different penguin species from the Palmer Archipelago in Antarctica.
1. Load the dataset using:
import seaborn as sns
penguins = sns.load_dataset("penguins")
Make sure matplotlib.pyplot is imported as well.
2. Create a figure with 2 side-by-side subplots.
3. For the leftmost plot, make a scatter plot using sns.scatterplot() and
add a title, axis labels and a legend.
4. For the leftmost plot, make a scatter plot using sns.scatterplot() and
add a title, axis labels and a legend. For the rightmost plot, make a
histogram using sns.histplot() to show the distribution of body mass
for all the penguins.

```python
import seaborn as sns
penguins = sns.load_dataset("penguins")
plt.figure()
x1 = penguins.loc[:, 'bill_length_mm']
y1 = penguins.loc[:, 'bill_depth_mm']
hist = penguins.loc[:, 'body_mass_g']
fig, axes = plt.subplots(nrows=1, ncols=2, figsize = (10,5))
sns.scatterplot(x=x1, y=y1, data=penguins, ax=axes[0], hue='sex')
axes[0].set_title('Penguin Bill Length vs. Bill Depth')
axes[0].set_xlabel('Bill Length (mm)')
axes[0].set_ylabel('Bill Depth (mm)')
sns.histplot(x=hist, data=penguins, ax=axes[1])
axes[1].set_title('Penguin Body Mass Histogram')
axes[1].set_xlabel('Body Mass (g)')
axes[1].set_ylabel('Count')
plt.show()
#A legend dosen't work here because only one set of points/data is being placed on each graph, as per the instructions.
```
